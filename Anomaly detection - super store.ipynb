```python
# Plot 3: Countplot of Ship Mode
plt.figure(figsize=(10, 6))
sns.countplot(x='Ship Mode', data=df)
plt.title('Countplot of Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 5: Lineplot of Quantity over Time (Order Date)
plt.figure(figsize=(14, 6))
sns.lineplot(x='Order Date', y='Quantity', data=df)
plt.title('Quantity over Time')
plt.xlabel('Order Date')
plt.ylabel('Quantity')
plt.show()
```

```python
# Plot 6: Pie chart of Segment Distribution
plt.figure(figsize=(5, 3))
df['Segment'].value_counts().plot.pie(autopct='%1.1f%%', startangle=50, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Segment Distribution')
plt.show()
```

```python
# Plot 7: Scatterplot of Discount vs. Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Profit', data=df, color='purple')
plt.title('Discount vs. Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

```python
# Plot 8: Pairplot of Selected Features
selected_features = ['Sales', 'Profit', 'Quantity', 'Discount']
sns.pairplot(df[selected_features])
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```

```python
# Plot 9: Violin Plot of Sales by Category
plt.figure(figsize=(12, 6))
sns.violinplot(x='Category', y='Sales', data=df)
plt.title('Violin Plot of Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
# Plot 10: Heatmap of Correlation Matrix (excluding non-numeric columns)
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_columns].corr()

plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

## Anamoly Detection using ISOLATION FOREST

```python
from sklearn.ensemble import IsolationForest

# Assuming 'Sales' and 'Profit' are relevant features for anomaly detection
features_for_anomaly_detection = ['Sales', 'Profit']

# Create a subset of the DataFrame with relevant features
X = df[features_for_anomaly_detection]

# Fit the Isolation Forest model
isolation_forest = IsolationForest(contamination=0.05)  # Adjust contamination based on your dataset
isolation_forest.fit(X)

# Predict anomalies (1 for normal, -1 for anomaly)
df['Anomaly'] = isolation_forest.predict(X)

# Display rows with anomalies
anomalies = df[df['Anomaly'] == -1]
print(anomalies)

# Visualize anomalies
plt.scatter(df['Sales'], df['Profit'], c=df['Anomaly'], cmap='viridis')
plt.title("Isolation Forest Anomaly Detection")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```
